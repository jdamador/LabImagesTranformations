# Balance de Color

## Objetivo
Escribir un programa simple que permita cambiar el balance de color de una imagen multiplicando cada valor de color por una constante especificadad por el usuario. 

## Ejemplo:

1. Importamos las librer√≠as.

```
from PIL import Image, ImageDraw
from matplotlib import pyplot as plt
```

2. Indicamos el valor de luminosidad que vamos a utilzar.

```
luminosity = 2
```

3. Cargamos la imagen que vamos a utilizar.

```
input_image = Image.open("./Resources/frutas.jpg")
input_pixels = input_image.load()
```

4. Crear una nueva imagen a partir de la existente.

```
output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)
```

5. Aplicamos el cambio de balance de color a cada pixel de la imagen.

```
# Generate image
for x in range(output_image.width):
    for y in range(output_image.height):
        r, g, b = input_pixels[x, y]
        r = int(r * luminosity)
        g = int(g * luminosity)
        b = int(b * luminosity)
        draw.point((x, y), (r, g, b))
```

6. Visualizar los resultados.

```
output_image.save("output.png")
plt.subplot(121),plt.imshow(input_image),plt.title('Input')
plt.subplot(122),plt.imshow(output_image),plt.title('Output')
plt.show()
```
